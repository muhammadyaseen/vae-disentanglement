```python
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
gen_images = 0

plt.rcParams['figure.figsize'] = (1.0, 1.0)

for r in range(5, 35):
    
    if gen_images > 3:
        break
    
    for h_raw in range(10, 40):

        for hole in range(6, 15):
        
            ball_r = r/30.0
            h = pow(ball_r,3)+h_raw/10.0 
            deep = hole/3.0
            
            ax = plt.gca()
    
            # water in cup 
            rect = plt.Rectangle(([3, 0]),5,5+h,color='lightskyblue')
            ax.add_artist(rect)
            ball = plt.Circle((5.5,+ball_r+0.5), ball_r, color = 'firebrick')
            ## cup
            left = plt.Polygon(([3, 0],[3, 19]), color = 'black', linewidth = 2)
            right_1 = plt.Polygon(([8, 0],[8, deep]), color = 'black', linewidth = 2)
            right_2 = plt.Polygon(([8, deep+0.4],[8, 19]), color = 'black', linewidth = 2)
            ax.add_artist(left)
            ax.add_artist(right_1)
            ax.add_artist(right_2)
            ax.add_artist(ball)
    
            #water line
            y = np.linspace(deep,0.5)
            epsilon = 0.01 * np.max([np.abs(np.random.randn(1)),1])
            x = np.sqrt(2*(0.98+epsilon)*h*(deep-y))+8
            x_max = x[-1]-8
            x_true = np.sqrt(2*(0.98)*h*(deep-0.5))
            plt.plot(x,y,color='lightskyblue',linewidth = 2)
    
            ##ground
            x = np.linspace(0,20,num=50)
            y = np.zeros(50)+0.2
            plt.plot(x,y,color='black',linewidth = 2)
            
            ax.set_xlim((0, 20))
            ax.set_ylim((0, 20))
    
            plt.axis('off')
            
            gen_images += 1
```

```python
fig, axes = plt.subplots(5,5) 
plt.rcParams['figure.figsize'] = (8.0, 8.0)
gen_images = 0

for i, ball_radius in enumerate(range(5, 10)):    
    for j, water_height in enumerate(range(10, 15)):
        gen_images += 1
        ball_radius_scaled = ball_radius / 30.0
        water_height_scaled = water_height / 10.0

        water_level = pow(ball_radius_scaled,3) + water_height_scaled/10.0 

        rect = plt.Rectangle((3., 0.), 5, 5 + water_level, color='lightskyblue')
        ball = plt.Circle((5.5, ball_radius_scaled + 0.5), ball_radius_scaled, color = 'firebrick')

        axes[i][j].add_artist(rect)
        axes[i][j].add_artist(ball)

        ##ground
        x = np.linspace(0,20,num=50)
        y = np.zeros(50)+0.2
        plt.plot(x,y,color='black',linewidth = 2)

        axes[i][j].set_xlim((0, 20))
        axes[i][j].set_ylim((0, 20))
        axes[i][j].axis('off')
```
